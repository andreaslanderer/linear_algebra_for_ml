# Solving Systems of Equations: Elimination

## Solving non-singular System of Linear Equations

The basic idea behind elimination is that we go from a system of equations to a **solved system** by isolating the variables. In order to achieve this we can manipulate the equations. There are several ways in which we can manipulate them:
* Multiplying by a constant:
$$ 
\begin{align*}
    a + b &= 10 \\
          &\times 7 \\
   7a + 7b &= 70
\end{align*} 
$$
* Adding/Subtracting two equations:
$$ 
\begin{align*}
    a + b &= 10 \\
          &+ \\
  2a + 3b &= 22 \\
          &= \\
  3a + 4b &= 32
\end{align*} 
$$

What we want to do is to manipulate one equation in a way that allows us to get rid of one of the coefficients and solve for the other one. Then we plug the solution into one of the equations and solve for the second one.

## Solving singular System of Equations

If we try to pursue the same strategy with a redundant, singular system we will get another result. Since the equations are basically identical, the equation itself is the solution. We get a **Degree of Freedom** (x). Let's assume we have the following equation:

$$ a + b = 10 $$

If we set one parameter to **x**, the other one is simply: **10 - x**

On the other hand, if we try to solve a contradictory, singular system using this approach the example will also be a contradiction, since there is no solution.

## Solving System of Equations with more variables

We can pursue a similar strategy when it comes to systems with more than two variables. Let's look at the following example:

$$ 
    a + b + 2c = 12  \\
    3a - 3b - c = 3 \\
    2a - b + 6c = 24
$$

First, we divide the second and third equation by the coefficient of $a$:

$$
    a + b + 2c = 12  \\
    a - b - \frac{1}{3}c = 1 \\
    a - \frac{1}{2}b + 3c = 12
$$

Then, we subtract the first equation from the others to eliminate $a$:

$$
    a + b + 2c = 12  \\
    -2b - \frac{7}{3}c = -11 \\
    -\frac{3}{2}b + c = 0
$$

Now, we have successfully isolated $a$ in the first equation, and we have a system of two equations with two variables that we can solve now:  

$$
    -2b - \frac{7}{3}c = -11 \\
    -\frac{3}{2}b + c = 0
$$

Let's start with solving for $b$ by eliminating $c$:

$$
    -\frac{6}{7}b - c = -\frac{33}{7} \\
    -\frac{3}{2}b + c = 0
$$

The next step is to add the two equations:

$$
\begin{align*}
    -\frac{33}{14}b &= -\frac{33}{7} \\
    b &= (-\frac{33}{7})\cdot(-\frac{14}{33}) \\
    b &= 2 
\end{align*}
$$

Now that we have a solution for $b$, let's determine the value of $c$:

$$
\begin{align*}
    -\frac{3}{2} \cdot 2 + c &= 0 \\
    c &= 3
\end{align*}
$$

And now we can easily determine $a$:

$$
\begin{align*}
    a + 2 + 2 \cdot 3 &= 12  \\
    a &= 4
\end{align*}
$$

## Matrix Row Reduction

Matrix row reduction, also known as **Gaussian elimination**, is a method for solving systems of linear equations. 
It offers several advantages:

1. **Simplicity**: It transforms a complex system of equations into a simpler one that is easier to solve, often reducing it to a form where the solution is immediately apparent.

2. **Versatility**: It can be used to solve systems of equations, find the rank of a matrix, determine the inverse of an invertible matrix, and solve linear programming problems.

3. **Algorithmic Efficiency**: It provides a systematic procedure that can be easily implemented in computer algorithms, making it suitable for computer-aided calculations with large systems of equations.

4. **Determining Consistency**: It can be used to determine whether a system of equations has no solution, one solution, or infinitely many solutions, based on the resulting row-echelon form.

5. **Flexibility**: It can be applied to any matrix, regardless of its dimensions, to simplify the matrix or solve associated equations.

Let's assume we have the following system of linear equations:
$$
    5a + b = 17 \\
    4a - 3b = 6 
$$

We can represent this system as matrices:


| Matrix                | Example                                            |
|-----------------------|----------------------------------------------------|
| Original Matrix       | $ \begin{pmatrix} 5 & 1 \\ 4 & -3 \end{pmatrix} $  |              
| Upper diagonal Matrix | $ \begin{pmatrix} 1 & 0.2 \\ 0 & 1 \end{pmatrix} $ |
| Diagonal Matrix       | $ \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} $   |
